```python
import pandas as pd
```

```python
data = pd.read_excel("/Users/natwarupadhyay/Downloads/BMWandFERRARI.xlsx")
data.head(5)
```

```python
count_short_words = lambda text: sum(1 for word in text.split() if len(word) <= 3 and word.isalpha())
data['Short Word Count'] = data['Brand Post'].apply(count_short_words)
print(data[['Short Word Count']])
```

```python
count_long_words = lambda text: sum(1 for word in text.split() if len(word) >= 10 and word.isalpha())
data['Long Word Count'] = data['Brand Post'].apply(count_long_words)
print(data[['Long Word Count']])
```

```python
count_hashtags = lambda text: len([word for word in text.split() if word.startswith('#')])
data['Hashtag Count'] = data['Brand Post'].apply(count_hashtags)
print(data[['Hashtag Count']])

```

```python
data['Brand Post'] = data['Brand Post'].str.lower()
data
```

```python
brand_post_counts = data['Brand Name'].value_counts()
brand_post_counts
```

```python
data['Post Length'] = data['Brand Post'].apply(lambda x: len(x.split()))
average_post_length = data['Post Length'].mean()
average_post_length
```

```python
max_post_length = data.groupby('Brand Name')['Post Length'].max()
min_post_length = data.groupby('Brand Name')['Post Length'].min()
max_post_length
```

```python
words = ' '.join(data['Brand Post']).split()
most_common_words = sorted(set(words), key=lambda word: words.count(word), reverse=True)[:3]
print(most_common_words)
```

```python
bmw_posts = ' '.join(data[data['Brand Name'] == 'BMW']['Brand Post']).lower()
ferrari_posts = ' '.join(data[data['Brand Name'] == 'FERRARI']['Brand Post']).lower()
bmw_words = set(bmw_posts.split())
ferrari_words = set(ferrari_posts.split())

unique_to_ferrari = ferrari_words - bmw_words
unique_to_bmw = bmw_words - ferrari_words

count_unique_to_ferrari = len(unique_to_ferrari)
count_unique_to_bmw = len(unique_to_bmw)

print(f"Unique words to BMW: {count_unique_to_bmw}")
print(f"Unique words to Ferrari: {count_unique_to_ferrari}")
```

```python
total_posts_bmw = len(data[data['Brand Name'] == 'BMW'])
total_posts_FERRARI = len(data[data['Brand Name'] == 'FERRARI'])
word_frequency = lambda text: pd.Series(text.split()).value_counts()
word_counts_bmw = data[data['Brand Name'] == 'BMW']['Brand Post'].apply(word_frequency)
word_counts_FERRARI = data[data['Brand Name'] == 'FERRARI']['Brand Post'].apply(word_frequency)

threshold_bmw = 0.3 * total_posts_bmw
threshold_FERRARI = 0.3 * total_posts_FERRARI


repeated_words_bmw = word_counts_bmw.columns[word_counts_bmw.apply(lambda x: (x >= threshold_bmw).sum() > 0)]
repeated_words_FERRARI = word_counts_FERRARI.columns[word_counts_FERRARI.apply(lambda x: (x >= threshold_FERRARI).sum() > 0)]


print("Words repeated in at least 30% of BMW's posts:", repeated_words_bmw)
print("Words repeated in at least 30% of FERRARI's posts:", repeated_words_FERRARI)
```

```python
average_length_bmw = data[data['Brand Name'] == 'BMW']['Brand Post'].apply(lambda x: len(x.split())).mean()
average_length_FERRARI = data[data['Brand Name'] == 'FERRARI']['Brand Post'].apply(lambda x: len(x.split())).mean()
print("Average post length for BMW:", average_length_bmw)
print("Average post length for FERRARI:", average_length_FERRARI)
```

```python
count_long_words = lambda text: sum(1 for word in text.split() if len(word) > 10)
count_long_words_b = lambda text: sum(1 for word in text.split() if len(word) > 2)

data['Long Word Count'] = data['Brand Post'].apply(count_long_words)
data['Long Word Count B'] = data['Brand Post'].apply(count_long_words_b)

for brand in data['Brand Name'].unique():
    brand_data = data[data['Brand Name'] == brand]
    total_posts = len(brand_data)
    
    posts_with_at_least_2_long_words = len(brand_data[brand_data['Long Word Count'] >= 2])
    posts_with_at_least_5_long_words = len(brand_data[brand_data['Long Word Count B'] >= 5])
    percentage_a = (posts_with_at_least_2_long_words / total_posts) * 100
    percentage_b = (posts_with_at_least_5_long_words / total_posts) * 100

    print(f"For {brand}:")
    print(f"(a) {percentage_a:.2f}% of total posts have at least 2 words with length greater than 10.")
    print(f"(b) {percentage_b:.2f}% of total posts have at least 5 words with length greater than 2.")
    print()
```

```python
brand_hashtag_counts = data.groupby('Brand Name')['Hashtag Count'].sum()
print(brand_hashtag_counts)
```

```python
import random

bmw_hashtags = data[data['Brand Name'] == 'BMW']['Brand Post'].str.findall(r'#\w+').sum()
ferrari_hashtags = data[data['Brand Name'] == 'FERRARI']['Brand Post'].str.findall(r'#\w+').sum()

bmw_hashtags = [hashtag[1:] for hashtag in bmw_hashtags]
ferrari_hashtags = [hashtag[1:] for hashtag in ferrari_hashtags]

random.shuffle(bmw_hashtags)
random.shuffle(ferrari_hashtags)

vs_outputs = []
for _ in range(10):
    selected_bmw_hashtag = random.choice(bmw_hashtags)
    selected_ferrari_hashtag = random.choice(ferrari_hashtags)
    vs_output = f'#{selected_bmw_hashtag} v/s {selected_ferrari_hashtag}'
    vs_outputs.append(vs_output)

# Viewing the output
print(vs_outputs)
```

```python
# Importing matplotlib library for visualization of distribution of post lengths for each brand
import matplotlib.pyplot as plt

# Calculating the length of each post
data['Post Length'] = data['Brand Post'].apply(lambda x: len(x.split()))

# Separating data for BMW and FERRARI
bmw_data = data[data['Brand Name'] == 'BMW']
ferrari_data = data[data['Brand Name'] == 'FERRARI']

# Creating histograms with post number on the x-axis and post length on the y-axis
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

ax1.hist(range(1, len(bmw_data) + 1), bins=len(bmw_data), weights=bmw_data['Post Length'], alpha=0.5, label='BMW', color='blue')
ax1.set_xlabel('Post Number')
ax1.set_ylabel('Post Length (Number of Words)')
ax1.set_title('Distribution of Post Lengths for BMW')
ax1.legend(loc='upper right')
ax1.grid(axis='y', linestyle='--', alpha=0.7)
ax1.set_axisbelow(True)

ax2.hist(range(1, len(ferrari_data) + 1), bins=len(ferrari_data), weights=ferrari_data['Post Length'], alpha=0.5, label='FERRARI', color='red')
ax2.set_xlabel('Post Number')
ax2.set_ylabel('Post Length (Number of Words)')
ax2.set_title('Distribution of Post Lengths for FERRARI')
ax2.legend(loc='upper right')
ax2.grid(axis='y', linestyle='--', alpha=0.7)
ax2.set_axisbelow(True)

# Viewing the output
plt.tight_layout()
plt.show()
```
